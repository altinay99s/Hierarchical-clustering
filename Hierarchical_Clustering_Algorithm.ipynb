#Алгоритм иерархической кластеризации
#(Hierarchical Clustering Algorithm)

```python
import numpy as np
import random
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

## Генерация точек

```python
n = 100
np.random.seed(0)
X = np.random.randn(n, 2)+np.random.randint(0, 10, (n, 2))
fig, ax = plt.subplots(figsize=(9, 7))
msize = 4.5

plt.plot (X[:, 0], X[:, 1], marker='o', linestyle='', markersize=msize)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

##Дендограма

```python
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

##Запуск иерархического алгоритма и вывод полученных класстеров

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
model.fit(X)
labels = model.labels_
fig, ax = plt.subplots(figsize=(9, 7))

plt.scatter(X[labels==0, 0], X[labels==0, 1], c='y', label='Класс 1')
plt.scatter(X[labels==1, 0], X[labels==1, 1], c='r', label='Класс 2')
plt.scatter(X[labels==2, 0], X[labels==2, 1], c='g', label='Класс 3')
plt.legend(loc='upper right', fontsize='7')
plt.show()
```
